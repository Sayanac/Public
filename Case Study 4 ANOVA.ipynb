# ● Check whether all the zones generate the same amount of sales

here One Way ANOVA test is used

setting up the hypothesis

Ho : All the Zones generate the same amount of sales.

Ha : All the Zones generate different amount of sales.

significance level=5%, α=0.05

```python
import scipy.stats as stats
fvalue,pvalue=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
print('fvalue:',fvalue.round(5))
print('pvalue:',pvalue.round(5))
```

```python
a=0.05
if a < pvalue :
    print('We fail to reject the NULL hypothesis')
else:
    print('Reject the NULL hypothesis')
```

Insight

from the one way ANOVA test it is clear that we should reject the null hypothesis.

# Therefor all the Zones generate different amount of sales
